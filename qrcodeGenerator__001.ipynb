```python
import qrcode
qr = qrcode.QRCode(
    version=2,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=15,
    border=3,
)
# Content
data = 'https://www.baidu.com/'
qr.add_data(data=data)
# Set color
qr.make(fit=True)
image = qr.make_image(fill_color='blue', back_color='white')
# Display QR code
image.show()
# Save it
image.save("baidu.jpg")
```

```python
from PIL import Image
from MyQR import myqr
import zxing

def createQR(u1, p1, s1):
  myqr.run(words = u1,
         version=1,
         picture = p1,
         colorized=True,
         save_name = s1)
  print("QR created")
  
reader = zxing.BarCodeReader()
def readQr(im):
  barcode = reader.decode(im)
  return barcode


u1 = "http://anadjangoapp.herokuapp.com/"
p1 = "/content/hr.png"
s1 = "/content/qr1.png"
createQR(u1, p1, s1)

print("Show QR")
im = Image.open(s1)
display(im)


print()
print("Read QR:")
text = readQr(s1)
print(text)
```
